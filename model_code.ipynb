## DATA CLEANING

```python
df=pd.read_csv("C:\\Users\\naren\\OneDrive\\Desktop\\udemy\\udemy_data\\Algerian_forest_fires_dataset.csv" )
df["region"]=0
df.iloc[0:122, -1]=0
df.iloc[124:247 , -1]=1                          # adding one more column region to help in merge two region data given 
df.region.astype(int)
df=df.dropna().reset_index(drop=True)
df
```

```python
df=df.drop(122).reset_index(drop=True)
df
```

```python
df.columns=['day', 'month', 'year', 'Temperature', 'RH', 'Ws', 'Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Classes', 'region']             
```

```python
df[["day","month","year","Temperature","RH","Ws"]]=df[["day","month","year","Temperature","RH","Ws"]].astype(int)
```

```python
df[["Rain","FFMC","DMC","DC","ISI","BUI","FWI"]]=df[["Rain","FFMC","DMC","DC","ISI","BUI","FWI"]].astype(float)
```

```python
df.info()
```

```python
df.drop(columns=["day","month","year"] , inplace=True)
```

```python
df
```

```python
df["Classes"]=np.where(df["Classes"].str.contains('not fire'), 0 ,1)
df
```

```python
df.corr()
```

```python
X=df.drop(columns="FWI")
y=df["FWI"]
```

# SPLITTING OF DATA

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.75, random_state=10)
```

## TRAINING OF DATA

```python
from sklearn.linear_model import ElasticNetCV
```

```python
model = ElasticNetCV(cv=5)
model.fit(X_train,y_train)
```

## ANALYSING ACCURACY

```python
from sklearn.metrics import r2_score
pre=model.predict(X_test)
result=r2_score(y_test,pre)
print("R2 SCORE:" , result)
```
